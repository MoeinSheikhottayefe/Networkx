In Python, there is a package for working with graphs and networks called NetworkX. This package is commonly imported as nx.

```python
import networkx as nx
```

To make a graph simple use the method nx.Graph() this would initialize an empty graph to work with.

```python
G = nx.Graph()
```

add_nodes_from() is method to add nodes to the graph G.

```python
G.add_nodes_from([1, 2, 3])
```

To add edges simply use the add_edges() method

```python
G.add_edge(1, 2)
```

For viewing the graph nodes you can use the method nodes()

```python
G.nodes()
```

And vise versa for the edges

```python
G.edges()
```

Node labels can be added using dictionaries. For example, G.nodes[2] accesses node 2, and a label can be assigned by setting G.nodes[2]['label'] = 'sky'.

```python
G.nodes[2]['label'] = 'sky'
```

To represent node labels, we enable the data argument for nodes. In this way, all nodes are represented as a dictionary of dictionaries, where the first element is the node identifier and the second element is a key–value dictionary containing the label and its corresponding value.

```python
G.nodes(data =True)
```

Also the networkx has its own way of drawing a graph using .draw() method which takes the graph G as an argument

```python
nx.draw(G)
```
