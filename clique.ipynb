```python
cimport networkx as nx

# 1. ساخت یک گراف نمونه (T) که دقیقاً 33 کلیک ماکسیمالِ 3تایی داشته باشد
T = nx.Graph()
# اضافه کردن 33 مثلث کاملاً مجزا به گراف (هر مثلث یک کلیک ماکسیمال سایز 3 است)
for i in range(33):
    # گره‌های i*3, i*3+1, i*3+2 را به هم متصل می‌کنیم تا یک مثلث تشکیل شود
    edges = [(i*3, i*3+1), (i*3+1, i*3+2), (i*3, i*3+2)]
    T.add_edges_from(edges)

# 2. تعریف تابع اصلی شما
def maximal_cliques_classic(G, size):
    """
    یافتن تمام کلیک‌های ماکسیمال در گراف G که دقیقاً دارای اندازه size هستند.
    """
    mcs = []
    for clique in nx.find_cliques(G):
        if len(clique) == size:
            mcs.append(clique)
    return mcs

# 3. تست کردن درستی کد
# اگر خروجی 33 نباشد، برنامه در این خط متوقف شده و خطای AssertionError می‌دهد
assert len(maximal_cliques_classic(T, 3)) == 33
print("کد با موفقیت اجرا شد و دقیقاً ۳۳ کلیک ماکسیمال پیدا شد.")
```

At firts, it is important to now the usage of deafualtdicts. If we pass the keyword(list) as the arguement and instatiate the defaultdict, it would be a good choice for organizing the neighbours.

```python
# 1. تعریف یک defaultdict که مقدار پیش‌فرض آن یک لیست خالی است
nodes_by_degree = defaultdict(list)

# 2. اطلاعات یک گره خاص را داریم:
degree = 3         # کلید ما (برچسب کشو)
node = "Node_A"    # مقدار ما (پرونده‌ای که می‌خواهیم در کشو بگذاریم)

# 3. عملیات اصلی:
nodes_by_degree[degree].append(node)
```

A common use of defaultdict is to build an adjacency list using edge lists:

```python
from collections import defaultdict

edges = [(0, 1), (0, 2), (1, 2)]

# ساخت گراف در یک چشم به هم زدن
adj_list = defaultdict(list)

for u, v in edges:
    adj_list[u].append(v)
    adj_list[v].append(u) # چون گراف بدون جهت است، برعکسش هم اضافه می‌شود

print(adj_list)
# خروجی: {0: [1, 2], 1: [0, 2], 2: [0, 1]}
```

```python
import networkx as nx
from collections import defaultdict

# 1. بارگذاری گراف استاندارد باشگاه کاراته زکری
# این گراف 34 گره (از 0 تا 33) و 78 یال دارد.
K = nx.karate_club_graph()

def analyze_all_maximal_cliques(G):
    """
    استخراج تمام کلیک‌های ماکسیمال و دسته‌بندی آن‌ها بر اساس اندازه (تعداد گره‌ها).
    """
    # استفاده از defaultdict برای ساخت یک دیکشنری که کلیدهای آن اندازه کلیک
    # و مقادیر آن، لیستی از کلیک‌های با آن اندازه است.
    cliques_by_size = defaultdict(list)

    # استخراج تمام کلیک‌های ماکسیمال (بدون فیلتر کردن سایز)
    for clique in nx.find_cliques(G):
        size = len(clique)
        # اضافه کردن کلیک استخراج شده به لیست مربوط به سایز خودش
        cliques_by_size[size].append(clique)

    return cliques_by_size

# 2. اجرای تحلیل روی گراف کاراته
clique_analysis = analyze_all_maximal_cliques(K)

# 3. نمایش گزارش آماری و ساختاری
print(f"Total number of maximal cliques: {sum(len(v) for v in clique_analysis.values())}")
print("-" * 50)

# مرتب‌سازی کلیدها (اندازه‌ها) از بزرگترین به کوچکترین کلیک
for size in sorted(clique_analysis.keys(), reverse=True):
    cliques = clique_analysis[size]
    print(f"Size {size}: Found {len(cliques)} maximal clique(s)")

    # چاپ کردن اعضای کلیک‌های بزرگ (مثلاً سایز 4 و 5) برای بررسی بیشتر
    if size >= 4:
        for idx, c in enumerate(cliques, 1):
            print(f"   -> Clique {idx}: Nodes {sorted(c)}")
    print("-" * 50)
```
